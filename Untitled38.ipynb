```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import read_csv
import seaborn as sns
from sklearn.preprocessing import StandardScaler
import datetime
import math
from numpy.random import choice
import random
from collections import namedtuple, deque
from tensorflow.keras.layers import MultiHeadAttention, LayerNormalization
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers
from tensorflow.keras.models import load_model
from tensorflow.keras.layers import Input, Dense, LSTM, Dropout, Attention, Concatenate, Flatten
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import Input, Dense, Dropout, Flatten, LayerNormalization, MultiHeadAttention, Embedding, TimeDistributed
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping

# Disable warnings
import warnings
warnings.filterwarnings('ignore')

# Ensure TensorFlow uses GPU if available
physical_devices = tf.config.experimental.list_physical_devices('GPU')
if physical_devices:
    try:
        for device in physical_devices:
            tf.config.experimental.set_memory_growth(device, True)
        print("GPU is available and memory growth is set.")
    except RuntimeError as e:
        print(e)
else:
    print("GPU not available, using CPU.")

# Load dataset
dataset = read_csv('Data\SP500_train.csv', index_col=0)
dataset.fillna(method='ffill', inplace=True)


# Prepare data
X = list(dataset["close"])
X = [float(x) for x in X]

validation_size = 0.2
train_size = int(len(X) * (1 - validation_size))
X_train, X_test = X[0:train_size], X[train_size:len(X)]

class Agent:
    def __init__(self, state_size, is_eval=False, model_name=""):
        self.state_size = state_size  # normalized previous days
        self.action_size = 3  # sit, buy, sell
        self.memory = deque(maxlen=1000)
        self.inventory = []
        self.model_name = model_name
        self.is_eval = is_eval

        self.gamma = 0.95
        self.epsilon = 1.0
        self.epsilon_min = 0.01
        self.epsilon_decay = 0.995
        self.model = self._model() if not is_eval else load_model(model_name)

    def _model(self):
        input_layer = Input(shape=(self.state_size, 1))

        # LSTM layers
        lstm_out = LSTM(units=32, return_sequences=True)(input_layer)
        lstm_out = LSTM(units=8, return_sequences=True)(lstm_out)
        lstm_out = LSTM(units=16, return_sequences=True)(lstm_out)
        
        # Fully connected layers
        dense_out = Dense(units=32, activation="relu")(lstm_out)
        dense_out = Dropout(0.2)(dense_out)
        dense_out = Dense(units=16, activation="relu")(dense_out)
        dense_out = Dropout(0.2)(dense_out)
        dense_out = Dense(units=4, activation="relu")(dense_out)

        # Output layer
        output_layer = Dense(self.action_size, activation="linear")(dense_out)

        # Define model
        model = models.Model(inputs=input_layer, outputs=output_layer)

        # Compile model
        model.compile(loss="mse", optimizer=Adam(learning_rate=0.001))

        return model

    def act(self, state):
        if not self.is_eval and np.random.rand() <= self.epsilon:
            return random.randrange(self.action_size)
        options = self.model.predict(state, verbose=0)
        return np.argmax(options[0])

    def expReplay(self, batch_size):
        mini_batch = random.sample(self.memory, batch_size)

        for state, action, reward, next_state, done in mini_batch:
            target = reward
            if not done:
                target = reward + self.gamma * np.amax(self.model.predict(next_state, verbose=0)[0])
            target_f = self.model.predict(state, verbose=0)
            target_f[0][action] = target
            early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
            history = self.model.fit(state, target_f, epochs=1000, batch_size=4096*32*32, verbose=0, validation_split=0.2, callbacks=[early_stopping])
        if self.epsilon > self.epsilon_min:
            self.epsilon *= self.epsilon_decay

def formatPrice(n):
    return ("-$" if n < 0 else "$") + "{0:.2f}".format(abs(n))

def getState(data, t, n):
    d = t - n + 1
    block = data[d:t + 1] if d >= 0 else -d * [data[0]] + data[0:t + 1]
    res = []
    for i in range(n - 1):
        res.append(block[i + 1] - block[i])
    return np.array([res]).reshape(-1, n - 1, 1)

def plot_behavior(data, states_buy, states_sell, total_profit):
    plt.figure(figsize=(10, 5))
    plt.plot(data, color='r', lw=2.)
    plt.plot(data, 'v', markersize=10, color='k', label='selling signal', markevery=states_sell)
    plt.plot(data, '^', markersize=10, color='m', label='buying signal', markevery=states_buy)
    plt.title(f'Total Profit: {formatPrice(total_profit)}')
    plt.legend()
    plt.show()

window_size = 60
batch_size = 8192  # Adjust the batch size
episode_count = 100000

agent = Agent(state_size=window_size)

with tf.device('/GPU:0'):
    for e in range(episode_count + 1):
        state = getState(X_train, 0, window_size + 1)
        total_profit = 0
        agent.inventory = []
        states_buy = []
        states_sell = []
        total_trades = 0
        profitable_trades = 0

        for t in range(len(X_train) - 1):
            action = agent.act(state)
            next_state = getState(X_train, t + 1, window_size + 1)
            reward = 0

            if action == 1:  # buy
                agent.inventory.append(X_train[t])
                states_buy.append(t)

            elif action == 2 and len(agent.inventory) > 0:  # sell
                bought_price = agent.inventory.pop(0)
                reward = X_train[t] - bought_price  # Allow negative rewards for losses
                total_profit += reward
                states_sell.append(t)
                total_trades += 1
                if reward > 0:
                    profitable_trades += 1

            done = t == len(X_train) - 2
            agent.memory.append((state, action, reward, next_state, done))
            state = next_state

            if done:
                if e % 1000 == 0:
                    print(f"Total Profit: {formatPrice(total_profit)}")
                    #plot_behavior(X_train, states_buy, states_sell, total_profit)
                    if total_trades > 0:
                        profitable_trade_ratio = profitable_trades / total_trades
                        print(f"Profitable Trades: {profitable_trades} / {total_trades} = {profitable_trade_ratio:.2f}")
                        total_trades = 0
                        profitable_trades = 0

            if len(agent.memory) > batch_size:
                agent.expReplay(batch_size)

        if e % 1000 == 0:
            agent.model.save(f"models/model_ep{e}.keras")

```

```python
# Function to calculate profit
def calculate_profit(predictions, actual_prices):
    cash = 10000  # Starting cash
    shares = 0
    buy_price = 0
    total_profit = 0
    trades = []

    for i, pred in enumerate(predictions):
        action = pred
        price = actual_prices[i]
        
        if action[1] == 1 and cash >= price and price!=0:  # Buy
            shares_to_buy = cash // price
            shares += shares_to_buy
            cash -= shares_to_buy * price
            buy_price = price
            trades.append((i, 'buy', price))
        elif action[2] == 1 and shares > 0:  # Sell
            cash += shares * price
            profit = (price - buy_price) * shares
            total_profit += profit
            shares = 0
            trades.append((i, 'sell', price))

    # Sell any remaining shares at the end
    if shares > 0:
        cash += shares * actual_prices[-1]
        profit = (actual_prices[-1] - buy_price) * shares
        total_profit += profit
        trades.append((len(actual_prices)-1, 'sell', actual_prices[-1]))

    return total_profit, trades

# Get actual prices for the test set
test_prices = df['close'].values

# Calculate profit
total_profit, trades = calculate_profit(predictions, test_prices)

print(f"Total predicted profit: ${total_profit:.2f}")
```
